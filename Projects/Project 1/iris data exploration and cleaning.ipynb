```python
# EXPLORING DATA ANALYSIS:
# Species Distributon:

df["species"].value_counts()

sns.countplot(x="species",data=df)

# Observation:
# All the species have equal numbers of Observations. By which they are balancing the dat sheet . 
```

```python
# SEPAL LENGTH AND SEPAL WIDTH:

import seaborn as sns

sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=df
)
#sns.title("Sepal Length and Sepal Width")
#sns.show()

#Explanation:
# Setosa species is clearly separable for others.
```
